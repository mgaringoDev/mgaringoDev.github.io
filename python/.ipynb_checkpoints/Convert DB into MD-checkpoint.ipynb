# Creating Blog Posts

```python
for index, row in db.iterrows():
    
    title = row["Title"].replace(" ","-")
    fileName = "{}-{}.md".format(row["Date"],title)
    fileLocation = "..\_posts\projects\{}".format(row["ImageFolderLocation"])
    fileSave = fileLocation + "\\" + fileName
    
    print("Now Creating File {}".format(fileName))
    
    f = open(fileSave,"w")
    
    f.write("---\n")
    f.write("layout: post\n")
    f.write("title: {}\n".format(row["Title"]))
    f.write("date: {}\n".format(row['Date']))
    f.write("tags: {} {}\n".format(row['Category'],row['Sub-Category']))    
    f.write("---\n")
    
    
    f.write("**Project Lifespan\\:** {}  \n".format(row["Year"]))
    f.write("<br>\n")
    
    # This was the working copy but it wasn't centering properly
    # f.write("![{}]({})\n".format(row["ProjectID"],row["ImageLocation"]))
    
    f.write("\n")
    f.write("{}".format(row["Description"]))
    
    f.close()
    print("Finsied")
```

# Creating Tags

```python
catList = []
for index, row in db.iterrows():
    itemCategoryString = "{} {}".format(row['Category'],row['Sub-Category'])   
    for word in itemCategoryString.split(" "):
        catList.append(word)   
```

```python
uniqueTags = np.unique(catList)
```

```python
for tag in uniqueTags:    
    fileName = "{}.md".format(tag)
    fileLocation = "..\pages\\tags\{}".format(fileName)
    fileSave = fileLocation 
    
    print("Now Creating File {}".format(fileName))
    
    f = open(fileSave,"w")
    
    f.write("---\n")
    f.write("layout: tag\n")
    f.write("title: {}\n".format(tag))        
    f.write("category: {}\n".format("tagPages"))
    f.write("---\n")
            
    f.close()
    print("Finsied")
```
